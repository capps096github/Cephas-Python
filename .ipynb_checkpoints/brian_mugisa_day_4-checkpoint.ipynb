```python
#Perform Operations on the data set in Python
import pandas as pd

# Read the data from the file
new_health_ds = pd.read_csv("health.csv").dropna()

# print the new data set
print(new_health_ds)

# show null values
new_health_ds.isnull().sum()

# dataset info
new_health_ds.info()

# show the first 5 rows
new_health_ds.head()

# show the last 5 rows
new_health_ds.tail()

# show the first 10 rows
new_health_ds.head(10)

# show the last 10 rows
new_health_ds.tail(10)

# describe the data set
new_health_ds.describe()

# show the data types
new_health_ds.dtypes

# get the column names
new_health_ds.columns

# get the index
new_health_ds.index

# get the values
new_health_ds.values

# get the shape
new_health_ds.shape

# get the size
new_health_ds.size

# get the mean
new_health_ds.mean()

# get the median
new_health_ds.median()

# get the mode
new_health_ds.mode()

# get the standard deviation
new_health_ds.std()

# get the variance
new_health_ds.var()

# get the sum
new_health_ds.sum()

# get the product
new_health_ds.prod()

# get the min
new_health_ds.min()

# get the max
new_health_ds.max()

# get the count
new_health_ds.count()

# get the cumsum
new_health_ds.cumsum()

# get the cumprod
new_health_ds.cumprod()

# fill in the missing values
new_health_ds.fillna(0)
```
