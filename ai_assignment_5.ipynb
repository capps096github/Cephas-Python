```python
# As an individual or team of two, build an end-to-end machine learning pipeline for your project by performing the following tasks:
# 1) Explore and clean data (Qn. 1)
# 2) Split data into train, validation and test sets
# 3) Fit any one initial model of your choice
# 4) Tune hyperparameters
# 5) Evaluate on validation set

import pandas as pd

# Read the data from the file
breast_cancer_ds = pd.read_csv("breast_cancer.csv")


# 7. Attribute Information: 
#    #  Attribute                     Domain
#    -- -----------------------------------------
#    1. Sample code number            id number
#    2. Clump Thickness               1 - 10
#    3. Uniformity of Cell Size       1 - 10
#    4. Uniformity of Cell Shape      1 - 10
#    5. Marginal Adhesion             1 - 10
#    6. Single Epithelial Cell Size   1 - 10
#    7. Bare Nuclei                   1 - 10
#    8. Bland Chromatin               1 - 10
#    9. Normal Nucleoli               1 - 10
#   10. Mitoses                       1 - 10
#   11. Class:                        (2 for benign, 4 for malignant)

# 1. Perform Data Cleaning by deleting the Sample code number column
# then also drop all the null values from the dataset
breast_cancer_ds.drop('Sample code number', axis=1, inplace=True)

# drop all the null values from the dataset
breast_cancer_ds.dropna(inplace=True)

# show the new data set
# print(breast_cancer_ds.to_string())


# 2) Split data into train, validation and test sets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np

# trainig data based on Class column
X_train = breast_cancer_ds.drop(columns=['Class'])
y = breast_cancer_ds['Class']

# split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_train, y, test_size=0.2, random_state=1)

# split the data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)

# 3) Fit any one initial model of your choice
# create a decision tree classifier
dtc = DecisionTreeClassifier()

# fit the model in the decision tree classifier but first convert the data into numpy arrays
dtc.fit(np.array(X_train), np.array(y_train))


# 4) Tune hyperparameters
# the dataset is a collection of medical records and the listed variables are attributes of the samples. In this case, the listed variables are not considered to be hyperparameters.

# 5) Evaluate on validation set
# calculate the accuracy score
accuracy = accuracy_score(y_val, dtc.predict(X_val))

# print the accuracy score
print("Accuracy: ", accuracy)
```
