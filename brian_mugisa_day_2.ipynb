```python
# create a list of cars of type list[str] and apply all the methods of the list while printing after each method

# list
cars = ["toyota", "honda", "bmw", "mercedes", "audi", "lexus", "volkswagen", "ford", "nissan", "mitsubishi"]

# print the list
print(cars)

# append a car to the list
cars.append("kia")
print(cars)

# insert a car to the list
cars.insert(0, "chevrolet")
print(cars)

# remove a car from the list
cars.remove("kia")
print(cars)

# pop a car from the list
cars.pop()
print(cars)

# sort the list
cars.sort()
print(cars)

# reverse the list
cars.reverse()
print(cars)

# count the number of cars in the list
print("We have " + str(len(cars)) + " cars in the list")


```

```python
# Python Iterators

# create a cars tuple of type tuple[str] and apply all the methods of the tuple while printing after each method

# tuple
cars = ("toyota", "honda", "bmw", "mercedes", "audi", "lexus", "volkswagen", "ford", "nissan", "mitsubishi")

# print the tuple
print("This tuple of cars = ",cars)

# iterators
cars_iterator = iter(cars)

# go to the next line
print("\n")

print("\n--- A tuple of cars in a loop ---\n")

print("a) With next() function\n")
# for each loop on the tuple
for car in cars:
    print("- {}".format( next(cars_iterator)))
    
    
# Looping without the next function
print("\nb) Without next() function\n")
for car in cars:
    print("* "+ car)
```

```python
# Python Iterators

# create a cars tuple of type tuple[str] and apply all the methods of the tuple while printing after each method

# tuple
cars = ("toyota", "honda", "bmw", "mercedes", "audi", "lexus", "volkswagen", "ford", "nissan", "mitsubishi")

# print the tuple
print("Old Tuple of cars")
for car in cars:
    print(" * "+ car)

# to add an item, first convert an item to a list, add an item, then convert the list into a tuple
y = list(cars)

# add mercedes to the new list at the end
y.append("mercedes")

# insert tesla at index 3
y.insert(3, "Tesla")

# re-assign the tuple with the new list value
cars = tuple(y)

# print the tuple
print("\n\nNew Tuple of cars")
for car in cars:
    print(" - "+ car)
```

```python
# Python Sets

# create a fruits set of type set[str] and apply all the methods of the set while printing after each method

# set
fruits = {"apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"}

# print the set
print("This set of fruits = ",fruits)

# loop through the set and print each item with *
print("\na) A set of fruits in a loop ---\n")
for fruit in fruits:
    print(" * "+ fruit)

# access an item in the set
print("\nb) Access an item in the set ---\n")

# check if the item is in the set
if "apple" in fruits:
    print(" * Yes, 'apple' is in the fruits set")

# access an item at a specific index
print("\nc) Access an item at a specific index ---\n")

# convert the set to a list and access the item at index 1
fruits_list = list(fruits)
print(" * The first item in the set is: " + fruits_list[0])
```

```python
# Python Dictionaries

# create a cars dictionary of type dict[str, str] and apply all the methods of the dictionary while printing after each method

# dictionary
cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

# print the dictionary
print("This dictionary of cars = ",cars)

# print the keys of the dictionary


# loop through the dictionary and print each item with *
print("\na) A dictionary of cars in a loop ---\n")
for car in cars:
    print(" * "+ car + " : " + str(cars[car]))

# access an item in the dictionary
print("\nb) Access an item in the dictionary ---\n")

# check if the item is in the dictionary
if "brand" in cars:
    print(" * Yes, 'brand' is one of the keys in the cars dictionary")
```

```python
 # Nested Loops in Python

flavors = ["chocolate", "vanilla", "strawberry"]
fruits = ["apple", "banana", "cherry"]

print("Flavors" +" "*(len("strawberry") - len("Flavors"))+ " | " + "Fruits\n-----------------------")
for x in flavors:
  for y in fruits:
      # check if the flavors length is less than the length of the chocolate
      # if it is less, then add spaces equivalent to the missing characters
      # else print the flavor and fruit
      if len(x) < len("strawberry"):
          print(x + " "*(len("strawberry") - len(x)) + " | " + y)
      else:
          print(x + " | " + y)
```
