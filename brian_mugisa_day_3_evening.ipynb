```python
# Pythton Objects and Classes

# class Person with name, age, and country
class Person:
    def __init__(self , name, age, country):
        self .name = name
        self .age = age
        self .country = country

    def details(self ):
        print("Hello, My name is " + self.name + " and I am " + str(self.age) + " years old.")

    def formatDetails(self ): 
        age_suffix = "years" if((self.age) > 1) else "year"
        print(f"\n* Formatted: {(self.name).upper()}'s Details:\n- Name: {self .name}\n- Age: {str(self.age)} {age_suffix} old.\n- Location (Country): {self.country}")
        username = (self.name).strip().lower().replace(" ","_").replace("c","x").replace("brian","official")
        print("- Username: ", "@" + username + "\n")

# Create an object of the class Person
cephas = Person("Cephas Brian", 2, "Uganda")
brianna = Person("Brianna", 1, "Uganda")
alexa = Person("Alexa", 1, "USA")

# Print the object
cephas.details()
brianna.details()
alexa.details()

# list of all persons
persons = [cephas, brianna, alexa]

# loop through the list of persons and print their details
for person in persons:
    person.formatDetails()
```

```python
# Python Inheritance

class Car:
    def __init__(self, make, manufacturer, yearOfManufacture, numberOfPassengers ):
        self.make =make
        self.manufacturer = manufacturer
        self.yearOfManufacture = yearOfManufacture
        self.numberOfPassengers = numberOfPassengers
        
    def details(self):
        print("Car Make: " + self.make + "\nManufacturer: " + self.manufacturer + "\nYear of Manufacture: " + str(self.yearOfManufacture) + "\nNumber of Passengers: " + str(self.numberOfPassengers) + "\n")
        
class Taxi(Car):
    def __init__(self,make, manufacturer, yearOfManufacture, numberOfPassengers, sound ):
        super().__init__(make, manufacturer, yearOfManufacture, numberOfPassengers )
        self.sound = sound
    
    def drive(self):
        print("The " + self.make + " goes like " + self.sound)
        
class Spacio(Car):
    def __init__(self,make, manufacturer, yearOfManufacture, numberOfPassengers, license ):
        super().__init__(make, manufacturer, yearOfManufacture, numberOfPassengers )
        self.license = license
   
    def showLicense(self):
        print("The " + self.make + " has a license number " + self.license)
        
class Rover(Car):
    pass


# Python Objects for the above classes
# a) Car class i.e Parent class
car = Car("Convertible", "Toyota", 2010, 4)
# print the details of the car
car.details()

taxi = Taxi("Toyota Taxi", "Toyota Uganda", 2015, 4, "Honk Honk"*3)
# print the details of the taxi and drive
taxi.drive()
taxi.details()


spacio = Spacio("VolksWagen Spacio", "VolksWagen Uganda", 2018, 4, "UAB 1234")
# print the details of the spacio and show license
spacio.showLicense()
spacio.details()

rover = Rover("Range Rover Sport", "Land Rover", 2019, 4)
# print the details of the rover
rover.details()


# a list of all cars
cars = [car, taxi, spacio, rover]

# loop through the list of cars and print their details
print(f"All Cars in the list of {type(cars)} and their details\n")
for car in cars:
    print("Car Make: ", car.make)
    print("Car Manufacturer: ", car.manufacturer)
    print("Car Year of Manufacture: ", car.yearOfManufacture)
    print("Car Number of Passengers: ", car.numberOfPassengers)
    print("Car Sound: ", car.sound if hasattr(car, "sound") else "N/A")
    print("Car License: ", car.license if hasattr(car, "license") else "N/A")
    print("\n")
```
