```python
import pandas as pd

# Read the data from the file
health_ds = pd.read_csv("health.csv")

# print the data on screen
# print(health_ds)

# show null values
# health_ds.isnull().sum()

# mean, median, mode
mean = health_ds["Calories"].mean()

# print the mean
# print(mean)

median = health_ds["Calories"].median()
mode = health_ds["Calories"].mode()


# menu
selected_index = input(f"What do you want to replace the null values with? \n  1. Mean ({mean}) \n  2. Median ({median}) \n  3. Mode of the Dataset ({mode}) \nInput accepted values: 1, 2, 3 \n")

# if statements
if selected_index == "1":
    print(f"Replaced null values with  Mean: {mean}\n ")
    # fill in the missing values with the mean
    health_ds.fillna(mean, inplace=True)
elif selected_index == "2":
    print(f"Replaced null values with  Median: {median}\n ")
    # fill in the missing values with the median
    health_ds.fillna(median, inplace=True)
  
elif selected_index == "3":
    print(f"Replaced null values with  Mode: {mode}\n ")
    # fill in the missing values with the mode
    health_ds.fillna(mode, inplace=True)

else:
    print("Invalid input")

# show the new data set
print(health_ds.to_string())
```

```python
# PIe Chart Example

import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Pie Charts in Python
import pandas as pd
import numpy as np # from pip install numpy
import seaborn as sns # from pip install seaborn
import matplotlib.pyplot as plt # from pip install matplotlib

# read the music data set csv
music_ds = pd.read_csv('dummy.csv')

# get the column names and change them to a list of labels
labels = tuple(music_ds["genre"].tolist())

# get the sizes of the data set from the age column and change them to a list of sizes integers
sizes =  music_ds['age'].tolist()

# get the explode only explode the first and second column and this is based on the length of the labels and convert it to a tuple
explode = (0.1, 0.1)+ tuple([0 for i in range(len(labels)-2)]) 

# Email: jeff.geoff.mis@gmail.com cc. jeff.geoff.cis@gmail.com

# colors from the data set colors column
colors =  music_ds['colors'].tolist()

# Colors can be
# b: blue.
# g: green.
# r: red.
# c: cyan.
# m: magenta.
# y: yellow.
# k: black.
# w: white.


fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels,shadow=True,autopct='%1.1f%%', colors =colors,
         startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# set the title
plt.title("Music Genre Pie Chart")

# set the size to 100 by 100
plt.figure(figsize=(100, 100))

# show the plot
plt.show()
```
