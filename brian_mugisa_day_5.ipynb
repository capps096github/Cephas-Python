```python
# Web Scrpping in Python
# site : https://keithgalli.github.io/web-scraping/webpage.html
import requests
import pandas as pd

from bs4 import BeautifulSoup as bs

url = "https://keithgalli.github.io/web-scraping/webpage.html"

# get the webpage
page = requests.get(url)

# parse the webpage
soup = bs(page.content, "html.parser")

# display the table from the webpage using pandas class= "hockey-stats"
hockey_table = soup.find(class_="hockey-stats")


# display the table from the webpage using pandas
print(hockey_table)
# get all text from the site
# print(soup.get_text())

# print(table.prettify())
```
