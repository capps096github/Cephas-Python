```python
# Create an ML model
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import dataset
x= np.random.normal(1, 100, 100)
y= np.random.normal(150, 40, 100)/x

# train x
train_x = x[:80]
# train y
train_y = y[:80]

# test x
test_x = x[:20]
# test y
test_y = y[:20]

# Visualize dataset
plt.scatter(x, y)
plt.show()

# Train the model
plt.scatter(train_x, train_y)
plt.show()

# test the model
plt.scatter(test_x, test_y)
plt.show()

# plot a line of best fit with the training data and numpy
poly = np.poly1d( np.polyfit(train_x, train_y, 1))
linegraph = np.linspace(0, 100, 100)

# plot the line of best fit
plt.scatter(train_x, train_y)
plt.plot(linegraph, poly(linegraph))
plt.show()
```
