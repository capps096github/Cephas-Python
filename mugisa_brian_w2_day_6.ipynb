```python
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

music_db = pd.read_csv('musicx.csv')

print(music_db.to_string())

# training data
X_train = music_db.drop(columns=['genre'])
y = music_db['genre']

# train test split
X_train, X_test, y_train, y_test = train_test_split(X_train, y, test_size=0.7)

# model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# predictions
predictions = model.predict(X_test)

# accuracy
score = accuracy_score(y_test, predictions)

print(score)

# 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pydot

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from six import StringIO

music_db = pd.read_csv('musicx.csv')

# print(music_db.to_string())

# training data
# everything else in the csv except genre
features = music_db.drop(columns=['genre'])

# genre column will give us the labels for the output
labels = music_db['genre']

# jeff.geoff.cis@gmail.com

# classifier from the decision tree
music_classifier = DecisionTreeClassifier()

# now we train the classifier with the features and labels
music_classifier.fit(features, labels)

# now we make predictions
# predictions = music_classifier.predict([[21, 1], [22, 0]])

# print(predictions)

# now we split the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)

# now we train the classifier with the training data
music_classifier.fit(X_train, y_train)

# now we make predictions with the testing data
predictions = music_classifier.predict(X_test)

# now we check the accuracy of the predictions
score = accuracy_score(y_test, predictions)

print(score)

# now we visualize the tree
tree.plot_tree(music_classifier)

# show the tree
plt.show()

# another way to visualize the tree
dot_data = "music_classifier.dot"

tree.export_graphviz(music_classifier, out_file=dot_data, feature_names=['age','gender'], class_names=sorted(labels.unique()), label='all', rounded=True, filled=True)

```
